```python
%%sql
CREATE TABLE Emp(eid int NOT NULL UNIQUE, ename varchar(200), age int, salary real, PRIMARY KEY(eid));
CREATE TABLE Works(eid int NOT NULL, did int NOT NULL, pct_time float, PRIMARY KEY(eid, did));
CREATE TABLE Dept(did int, dname varchar(200), budget real, managerid int, FOREIGN KEY (managerid) REFERENCES Emp(eid));
```

```python
#worked in at least 1 dept
%%sql
SELECT E.eid
FROM Emp E, Works W1, Works W2
WHERE W1.eid = E.eid and W2.eid = E.eid and W1.did <> W2.did;
```

## 1.1 Managerid's of those who manage depts with budgets > 1MM

```python
%%sql
SELECT DISTINCT D.managerid FROM Dept D WHERE D.budget >= 1000000;
```

## 1.2 Employees that work in more than one dept

```python
%%sql
SELECT e.ename
FROM Emp e, Works w
WHERE e.eid=w.eid
GROUP BY e.eid
HAVING COUNT(e.eid)>=2;
```

## 1.3 Managers who manage depts with smallest budget

```python
%%sql
SELECT DISTINCT E.ename
FROM Emp E
WHERE E.eid 
IN (
    SELECT D.managerid 
    FROM Dept D 
    WHERE D.budget = (SELECT MIN (D.budget) FROM Dept D)
    );
```

## 1.4 Employee name for employees that works in most depts

```python
%%sql
SELECT e.ename 
FROM Emp e, Works w
WHERE e.eid=w.eid
GROUP BY e.eid
HAVING COUNT(e.eid)>=2;
```
